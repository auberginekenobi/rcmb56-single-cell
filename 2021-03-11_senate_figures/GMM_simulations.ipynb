```python
# example of a bimodal constructed from two gaussian processes
import numpy as np
from numpy import hstack
from numpy.random import normal
from matplotlib import pyplot

def simulate_ecDNA_counts(n_cells,mu_ec=30,mu_l=20, sigma=5):
    '''
    n_cells: list of tuples [(ecdna_count,n_cells)] - number of ecDNAs per cell, number of cells.
    '''
    flag = False
    for i in range(len(n_cells)):
        ec_ct, cell_ct = n_cells[i]
        X = normal(loc=mu_l, scale=sigma, size = cell_ct)
        for j in range(ec_ct):
            X += normal(loc=mu_ec, scale = sigma, size = cell_ct)
        if not flag:
            joint = X
            flag = True
        else:
            joint = hstack((joint,X))
    pyplot.hist(joint, bins=50, density=True)
    pyplot.show()

```

```python
sim1 = [(0,3000),(1,7000)]
simulate_ecDNA_counts(sim1)
```

```python
sim2 = [(0,3000),(1,7000),(2,10000),(3,12000),(4,10000),(5,4000)]
simulate_ecDNA_counts(sim2)
```
