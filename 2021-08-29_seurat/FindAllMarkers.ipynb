```r
library(dplyr)
library(ggplot2)
library(EnsDb.Hsapiens.v86)
library(Seurat)
library(Signac)
library(repr)
library(limma)
```

```r
sample_name <- "rcmb56-ht"
#sample_name <- "rcmb56-pdx"

if(sample_name == 'rcmb56-ht'){
    #rds <- 'rcmb56-ht_seuratobjdfclean.rds'
    rds <- file.path("out","rcmb56-ht_seuratobj.rds")
} else if(sample_name == 'rcmb56-pdx'){
    rds <- 'rcmb56-pdx_xenocelldfcleanseuratobj.rds'
}

seuratobj <- readRDS(rds)
seuratobj
```

```r
# Find differential genes
DefaultAssay(seuratobj) <- "SCT"
markers <- FindAllMarkers(seuratobj)
```

```r
markers
```

```r
markers[markers$gene %in% c('KMT2E','DNTTIP2','ETV1','GLI2'),]
```

```r
markers$delta <- markers$pct.1 - markers$pct.2
```

```r
attach(markers)
markers <- markers[ which(avg_log2FC > 0 & delta > .1),]
detach(markers)
```

```r
write.table(markers,file = paste0("out/",sample_name,'_FindAllMarkers.tsv'),sep='\t')
```
