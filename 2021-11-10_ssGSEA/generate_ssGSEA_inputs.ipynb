```r
library(dplyr)
library(ggplot2)
library(EnsDb.Hsapiens.v86)
library(Seurat)
library(Signac)
library(tibble)
library(readr)
```

```r
sample_name <- "rcmb56-ht"
#sample_name <- "rcmb56-pdx"
seurat_dir="../2021-08-29_seurat/"
seuratobj <- readRDS(paste0(seurat_dir,sample_name,"_seuratobj.rds"))
DefaultAssay(seuratobj) <- "SCT"
seuratobj
```

```r
m <- GetAssayData(seuratobj, slot="scale.data")
file=paste0(sample_name,"_scale.data.tsv")
write_gct <- function(datatable,filename){
    df = as_tibble(datatable)
    lines = c("#1.2",paste0(dim(datatable)[1],"\t",dim(datatable)[2]))
    writeLines(lines,filename)
    df <- add_column(df,NAME=rownames(datatable),Descripion="",.before=1)
    write.table(df,file=filename,quote=FALSE,sep='\t',append=TRUE,row.names=FALSE)
}
write_gct(m,file)
```
