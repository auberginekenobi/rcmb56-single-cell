```r
# Plot marker genes that we know are cluster-specific, and also represent known cell types.
library(Seurat)
library(Signac)
```

```r
seuratobj<- readRDS('../2021-08-29_seurat/rcmb56-ht_seuratobj.rds')
DefaultAssay(seuratobj) <- 'SCT'
```

```r
FeaturePlot(seuratobj,reduction="wnn.umap",features='MBP') # oligodendrocyte marker
FeaturePlot(seuratobj,reduction="wnn.umap",features='CD86') # marker for antigen-presenting cells inc. macrophage, dendritic, langerhans cells
FeaturePlot(seuratobj,reduction="wnn.umap",features='PTPRC',max.cutoff=5) # = CD45, marker for all differentiated immune cells
FeaturePlot(seuratobj,reduction="wnn.umap",features='VIM') # Vimentin, EMT marker.
```

```r
FeaturePlot(seuratobj,reduction="wnn.umap",features='NDRG2') # Astrocyte marker
FeaturePlot(seuratobj,reduction="wnn.umap",features='TNR',max.cutoff=20) # OPC marker
FeaturePlot(seuratobj,reduction="wnn.umap",features='PDGFRA') # OPC marker
```

```r
FeaturePlot(seuratobj,reduction="wnn.umap",features='GLI2',slot='scale.data')
FeaturePlot(seuratobj,reduction="wnn.umap",features='KMT2E',slot='scale.data')
FeaturePlot(seuratobj,reduction="wnn.umap",features='DNTTIP2',slot='scale.data')
FeaturePlot(seuratobj,reduction="wnn.umap",features='ETV1',slot='scale.data')
```
