```python
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
def get_known_markers(source):
    if source == 'literature':
        annot_file="../gene annotations neuron.xlsx"
        annot=pd.read_excel(annot_file,skiprows=2,index_col='gene')
    elif source == 'specific':
        annot_file="McKenzie_2018_S1.xlsx"
        annot=pd.read_excel(annot_file,skiprows=2,index_col='gene',sheet_name='top_all_specificity')
    else:
        raise ValueError("source must be 'literature' or 'specific'")
    return annot
def get_seurat_markers():
    markers_file="../2021-08-29_seurat/rcmb56-ht_markers.tsv"
    markers=pd.read_csv(markers_file,sep='\t')
    markers=markers.set_index('gene')
    return markers
def find_known_markers(source):
    annot=get_known_markers(source)
    markers=get_seurat_markers()
    markers=markers.merge(annot,how='inner',left_index=True,right_index=True)
    markers=markers.sort_values('cluster')
    null_count_distribution = annot.groupby('Celltype').count().iloc[:,1]
    print(null_count_distribution)
    for i in markers.cluster.unique():
        df = markers[markers.cluster == i]
        counts = df.groupby('Celltype').count().iloc[:,1]
        celltype=counts.index[counts.argmax()]
        print("cluster {}: {}".format(i,celltype))
        if source == 'literature':
            top_markers=df[(df.Celltype==celltype)].sort_values('Cell_type_mentions',ascending=False)
            top_markers=top_markers.index.tolist()[:5]
            print("Candidate markers: {}".format(', '.join(top_markers)))
        #print(df)
    return df
df=find_known_markers('literature')
print("#################################")
find_known_markers('specific')
```
