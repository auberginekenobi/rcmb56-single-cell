```python
import os
os.chdir('/mnt/c/Users/ochapman/Documents/Mesirov/scRNA+ATAC/2021-07-22_ecdna-featurecounts')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
counts = pd.read_csv('RCMB56-ht.tsv',sep='\t')

def plot_coverage_1():
    df = counts.copy()
    df["ecDNA1_depth_normalized"]=df["ecDNA1"]/df["chr2"]*100+1
    ax = sns.histplot(data=df,x='ecDNA1_depth_normalized',
                      log_scale=True)
    plt.savefig("ecDNA1.png")
    plt.savefig("ecDNA1.svg")
    return df
df = plot_coverage_1()
    
```
